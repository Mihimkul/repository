```python
import pandas as pd
df = pd.read_excel("Local_Electronic_Sales.xlsx")
df = df.drop(12, axis=0)
df['Month'] = pd.to_datetime(df['Month'])
df
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
plt.plot(df['Month'], df['TVs'], color='red', marker='o', label='TVs')

plt.title("Продажи телевизоров в 2022 году")
plt.xlabel("Месяцы")
plt.ylabel("Количество единиц")
plt.grid()
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot(df['Month'], df['TVs'], color='red', marker='o', label='TVs')
plt.plot(df['Month'], df['Laptops'], color='green', marker='s', label='Lapt')

plt.title("Продажи телевизоров в 2022 году")
plt.xlabel("Месяцы")
plt.ylabel("Количество единиц")
plt.grid()
plt.legend()
plt.show()
```
